```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='darkgrid')
```

```python
tips = sns.load_dataset('tips')
# scatterplot()
sns.relplot(x='total_bill', y='tip', data=tips)
```

```python
# hue semantic
# another dimension can be added
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
# Different style 
sns.relplot(x='total_bill', y='tip', hue='smoker', data=tips)
```

```python
# 4 variables
# categorical
sns.relplot(x='total_bill',y='tip',hue='smoker',style='time',data=tips)
```

```python
#
sns.relplot(x='total_bill',y='tip',hue='size',data=tips)
```

```python
# change colour palette
sns.relplot(x="total_bill", y="tip", hue="size", 
            palette="ch:r=-.5,l=.75", data=tips);
```

```python
sns.relplot(x='total_bill',y='tip',size='size',data=tips)
```

```python
#Normalize
sns.relplot(x='total_bill',y='tip',size='size',sizes=(15,200),data=tips)
```

# Continuity with the line plots

```python
#lineplot
df = pd.DataFrame(dict(time=np.arange(500),
                       value=np.random.randn(500).cumsum()))

g = sns.relplot(x='time',y='value',kind='line',data=df)
# sort the data by x value
g.fig.autofmt_xdate()
```

# Aggregation and representing uncertainty

```python
fmri = sns.load_dataset("fmri")
fmri.head()
```

```python
# same value of x
sns.relplot(x='timepoint',y='signal',kind='line',data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", ci=None, kind="line", data=fmri);
```

```python
sns.relplot(x='timepoint',y='signal',kind='line',ci='sd',data=fmri)
```

```python
sns.relplot(x='timepoint',y='signal',estimator=None,kind='line',data=fmri)
```

# Plotting subsets of data with semantic mappings

```python
# event = 2 colours
sns.relplot(x='timepoint',y='signal',hue='event',kind='line',data=fmri)
```

```python
# dashes
sns.relplot(x='timepoint',y='signal',hue='region',
            style='event',
            kind='line',data=fmri)
```

```python
# identify subset by markers
sns.relplot(x='timepoint',y='signal',hue='region',style='event',
           dashes=False, markers=True, kind='line',data=fmri)
```

```python
sns.relplot(x="timepoint", y="signal", hue="event", style="event",
kind="line", data=fmri);
```

```python
# repeated measures data, can also plot each sampling unit separately 

sns.relplot(x='timepoint', y='signal', hue='region',
           units='subject', estimator=None,
           kind='line',data=fmri.query("event == 'stim'"))
```

```python
dots = sns.load_dataset('dots').query('align == "dots"')
dots.head()
```

```python
# poor color scale
sns.relplot(x='time', y='firing_rate',
           hue='coherence', style='choice',
           kind='line', data=dots)
```

```python
# better color scale
palette = sns.cubehelix_palette(light=.8, n_colors=6)
sns.relplot(x='time', y='firing_rate',
           hue='coherence', style='choice',
           palette=palette,
           kind='line',data=dots)
```

```python
# normalize colormap
from matplotlib.colors import LogNorm
palette = sns.cubehelix_palette(light=.7, n_colors=6)
sns.relplot(x='time', y='firing_rate',
           hue='coherence',style='choice',
           hue_norm=LogNorm(),
           kind='line', data=dots)
```

```python
sns.relplot(x='time', y='firing_rate',
           size='coherence', style='choice',
           kind='line', data=dots)
```

```python
# width of line
sns.relplot(x='time', y='firing_rate',
           hue='coherence', size='choice',
           palette=palette,
           kind='line', data=dots)
```

# Plotting with date data

```python
df = pd.DataFrame(dict(time=pd.date_range("2017-1-1", periods=500),
                       value=np.random.randn(500).cumsum()))
df.head()
```

```python
g = sns.relplot(x="time", y="value", kind="line", data=df)
g.fig.autofmt_xdate()
```
